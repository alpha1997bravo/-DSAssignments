```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('Q9_a (1).csv') #loading data
```

```python
data.head() #it will 1st five columns
```

```python
data=data.set_index('Index')
```

```python
sns.displot(x='speed',data=data,kde=True,color='green')
```

```python
data['speed'].skew()
```

```python
# Tail of the distribution is skewed to the left.
```

```python
data['speed'].kurtosis()
```

```python
# The peakness of the distribution is light.
```

```python
sns.displot(x='dist', data=data, kde=True,color='green')
```

```python
data['dist'].skew()
```

```python
# Tail of the distribution is skewed to the Right.
```

```python
data['dist'].kurtosis()
```
